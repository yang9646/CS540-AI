```python
import csv
import numpy as np
import matplotlib.pyplot as plt
```

```python
# takes in a string with a path to a CSV file formatted as in the link above, 
# and returns the first 20 data points 
# (without the Generation and Legendary columns 
# but retaining all other columns) in a single structure.
def load_data(filepath):
    file = csv.DictReader(open(filepath))
    data = []
    temp = {}
    count = 0
    for row in file:
        if count == 20:
            return data
        count += 1
        temp = {'#': int(row['#']), 'name': row['Name'], 'Type 1':row['Type 1'], 'Type 2':row['Type 2'], 'Total':int(row['Total']), 
                'HP':int(row['HP']), 'Attack':int(row['Attack']), 'Defense':int(row['Defense']), 
               'Sp. Atk':int(row['Sp. Atk']), 'Sp. Def':int(row['Sp. Def']), 'Speed':int(row['Speed'])}
        data.append(temp)

    # row length = 13
                
    return data
```

```python
pokemons = load_data('pokemon.csv')
len(pokemons)
```

```python
# takes in one row from the data loaded from the previous function, 
# calculates the corresponding x, y values for that Pokemon as specified above, 
# and returns them in a single structure.
def caculate_x_y(stats):
    # x = total offensive strength
    # attack + Sp. atk + speed
    # y = total defensive strength
    # Defense + Sp.def + HP
    
    x = stats['Attack'] + stats['Sp. Atk'] + stats['Speed']
    y = stats['Defense'] + stats['Sp. Def'] + stats['HP']
    return x, y
```
